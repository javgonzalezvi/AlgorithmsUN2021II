## Time complexity analysis of MAX-HEAPIFY

$$T(n) \leq T(2n/3) + \Theta(1)  $$
$$T(n) = O(lg \  n) $$
\
Each call to **MAX-HEAPIFY** costs $O(lg \ n)$ time, and **BUILD-MAX-
HEAP** makes $O(n)$ such calls. Thus, the running time is $O(n \ lg \ n)$ This
upper bound, though correct, is not asymptotically tight.
\
\
We can derive a tighter bound by observing that the time for **MAX-HEAPIFY** to
run at a node varies with the height of the node in the tree, and the heights of most nodes are small. Our tighter analysis relies on the properties that an n-element heap has height $\lfloor lg \ n \rfloor$  and at most $\lceil n/2^{h+1}$
nodes of any height $h$.

# Experimental analysis of the Heap Sort execution time 

```
def parent(i): 
    return int(i/2)

def left(i):
    return 2*i

def right(i):
    return 2*i + 1

def max_heap(A, i):
    
    global heap_size
    l = left(i) 
    r = right(i) 

    if l <= heap_size and A[l] > A[i]:
        largest = l
    else:
        largest = i
    if r <= heap_size and A[r] > A[largest]:
        largest = r
    if largest != i:
        A[i], A[largest] = A[largest], A[i]
        max_heap(A, largest)


def build_max_heap(A):
    global heap_size
    heap_size = len(A)-1
    for i in range( int(len(A)/2)+1, 0, -1):
        max_heap(A,i-1) 

def heapsort(A):
    global heap_size
    build_max_heap(A)
    for i in range(len(A), 1, -1):
        A[0], A[i-1] = A[i-1], A[0]
        heap_size = heap_size - 1
        max_heap(A, 0)
```

# Heap sort  time complexity histograms for permutations of $n$ different elements counting steps 


 


$$t_a(n) = n + (n-1) \ log \ n + O(1) = \Theta(n \ log \ n)$$

```
import matplotlib.pyplot as plt
import numpy as np
import math

n = np.arange(1,21,1)
m = [math.log(i,2)*i for i in n]

plt.plot(n, m, label = 'Heap sort time complexity $n \ log \ n$', color="orange")
plt.xlabel('Array size')
plt.ylabel('Steps')
plt.legend(loc = 'upper left')
```

```
def parent(i): 
    global steps
    return int(i/2)
    steps = steps+1

def left(i):
    global steps
    return 2*i
    steps = steps+1

def right(i):
    global steps
    return 2*i + 1
    steps = steps+1


def max_heap(A, i):
    global heap_size
    global steps
    l = left(i) 
    r = right(i) 
    if l <= heap_size and A[l] > A[i]:
        largest = l
        steps = steps+1
    else:
        largest = i
        steps = steps+1
    steps = steps+1
    if r <= heap_size and A[r] > A[largest]:
        largest = r
        steps = steps+1
    steps = steps+1
    if largest != i:
        A[i], A[largest] = A[largest], A[i]
        max_heap(A, largest)
        steps = steps+1
    steps = steps+1

def build_max_heap(A):
    global heap_size
    global steps 
    heap_size = len(A)-1
    steps = steps+1
    for i in range(int(len(A)/2)+1, 0, -1):
        max_heap(A,i-1) 
        steps = steps+2


def heap_steps(A):
    global heap_size
    global steps 
    steps = 0
    build_max_heap(A)
    for i in range(len(A), 1, -1):
        A[0], A[i-1] = A[i-1], A[0]
        heap_size = heap_size - 1
        max_heap(A, 0)
        steps = steps+4
    return steps 
```

```
A = [1,2,3,4,5,6,7,8]
print('original',A)
print("pasos",heap_steps(A))
print(A,'\n')

A = [6,7,2,5,4,3,8,1]
print('original',A)
print("pasos",heap_steps(A))
print(A,'\n')

A = [8,7,6,5,4,3,2,1]
print('original',A)
print("pasos",heap_steps(A))
print(A)
```

Generate all permutations without repetitions of size *n* (for small *n*), plot the "exact" time  distribution (number of steps), and print the media and the standar deviation

```
from itertools import permutations
import matplotlib.pyplot as plt
import numpy as np
import math

n=3
X = range(n)
times = []

for P in permutations(X):
  t = heap_steps(list(P)) 
  times.append(t)  
b = np.array(range(min(times)-1, max(times)+1))+0.5
plt.hist(times, bins = b )
plt.title("Heap Sort")
plt.xlabel("Steps")
plt.ylabel("Frequency")
plt.show()
    
print("n",n)
print("numero de permutaciones n! =",math.factorial(n)) 
print("exp min =",min(times))

print("exp max =",max(times))

print("exp mean =",np.mean(times))

print("stdev",np.std(times))
```
